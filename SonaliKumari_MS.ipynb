**Loading the Dataset:**

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

pandas : used for data manipulation and analysis

Seaborn :  A stastical data visualization library based on matplotlib.

Matplotlib.pyplot :  used for craeting static, animated , and interactive visualizations.

numpy:  a fundamental package for scientific computing with python.


```python
# Load the dataset
df = sns.load_dataset('iris')
```

loads the famous iris dataset from seaborn's built-in datasets.

```python
# Display the first 5 rows of the DataFrame
print("First 5 rows of the dataset:")
print(df.head())
```

Print the first 5 roes to get an initial look at the datasets.

```python
# Display the summary statistics of the dataset
print("\nSummary statistics of the dataset:")
print(df.describe())
```

Provides summary statstics (count, mean, std, min) for numerical columns.

**Data Cleaning**

```python
# Data Cleaning
# Check for missing values
print("\nMissing values in the dataset:")
print(df.isnull().sum())
```

checks for any missing values in the dataset.

```python
# Ensure all column names are in a consistent format
df.columns = df.columns.str.lower()
print("\nColumn names after formatting:")
print(df.columns)
```

converts all column names to lowercase for consistency.

**Data Transformation**

```python
# Data Transformation
# Add a new column sepal_area
df['sepal_area'] = df['sepal_length'] * df['sepal_width']
```

Add two new columns'sepal_area' and 'petal_area' by multiplying the respective lengths and widths.

```python
# Add a new column petal_area
df['petal_area'] = df['petal_length'] * df['petal_width']
```

normalizes the specific columns to a 0-1 range.

```python
# Normalize the values in sepal_length, sepal_width, petal_length, and petal_width
df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']] = \
    df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].apply(lambda x: (x - x.min()) / (x.max() - x.min()))


```

shows the transformed datasets.

```python

print("\nFirst 5 rows of the dataset after transformation:")
print(df.head())
```

**Data Aggregation and Grouping**

Groups the dataser by species and calculates mean, median, and standard devaiation for each group.

```python
# Data Aggregation and Grouping
# Group by species and calculate mean, median, and standard deviation
grouped_stats = df.groupby('species').agg(['mean', 'median', 'std'])
print("\nGrouped statistics by species:")
print(grouped_stats)
```

creates a pivot table showing the mean sepal and petal area for each species.

A PivotTable is a powerful tool to calculate, summarize, and analyze data that lets you see comparisons, patterns, and trends in your data.

```python
# Create a pivot table for mean sepal_area and petal_area for each species
pivot_table = df.pivot_table(values=['sepal_area', 'petal_area'], index='species', aggfunc='mean')
print("\nPivot table showing mean sepal_area and petal_area for each species:")
print(pivot_table)
```

**Data Visualization**

Plots the distribuiition of sepal length,colored by species

```python
# Data Visualization
# Plot the distribution of sepal_length for each species
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='sepal_length', hue='species', kde=True)
plt.title('Distribution of Sepal Length by Species')
plt.show()
```

craetes a scatter plot of sepal length vs. sepal width , colored by species.

```python
# Scatter plot of sepal_length vs. sepal_width colored by species
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs. Sepal Width by Species')
plt.show()
```

Plots a heatmap of the correlation matrix to show relationships between features.

```python
# Heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 6))
# Exclude the 'species' column for correlation matrix calculation
correlation_matrix = df.drop(columns=['species']).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of the Iris Dataset')
plt.show()
```

**Advanced Data Analysis**

Retrieves the top 10 rows with the highest sepal area.

```python
# Advanced Data Analysis
# Identify the top 10 rows with the highest sepal_area
top_10_sepal_area = df.nlargest(10, 'sepal_area')
print("\nTop 10 rows with the highest sepal_area:")
print(top_10_sepal_area)
```

filters rows where petal length is greater than the median petal length.

```python
# Filter the dataset to only include rows where petal_length is greater than the median petal_length
median_petal_length = df['petal_length'].median()
filtered_df = df[df['petal_length'] > median_petal_length]
print("\nRows where petal_length is greater than the median petal_length:")
print(filtered_df)
```

plots a scatter plot showing the realtionship between petal length and petal width for each species.

```python
# Additional analysis: Relationship between petal_length and petal_width for each species
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
plt.title('Petal Length vs. Petal Width by Species')
plt.show()
```

saves the dataframe to a csv file.

```python
# Save the original dataset in CSV format
df.to_csv('iris_dataset.csv', index=False)
```
